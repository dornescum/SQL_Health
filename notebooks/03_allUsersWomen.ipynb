Select all users

```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
import pandas as pd

load_dotenv()

engine = create_engine(
    f"mysql+pymysql://{os.getenv('DB_USER')}:{os.getenv('DB_PASS')}"
    f"@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)
```

```python
df = pd.read_sql("SELECT id, name, surname, sex, country, town, role_id, is_active FROM users where sex =2", engine)
df.head()
# print(df.count())
print(df)
```
