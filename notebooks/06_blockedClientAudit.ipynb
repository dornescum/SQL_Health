# 06 — Blocked client audit

Self-join on `users.blocked_by` to pair each blocked client with the admin who blocked them. Mirrors `docs/Client_Querires.md` §1.3.

Contains client names/emails — for local exploration only, not for the README highlights.

```python
import pandas as pd
from db import get_engine

engine = get_engine()
```

```python
query = """
SELECT
  blocked.id AS client_id,
  blocked.name,
  blocked.surname,
  blocked.email,
  blocked.blocked_reason,
  blocked.blocked_at,
  admin.name AS blocked_by_admin
FROM users blocked
LEFT JOIN users admin ON admin.id = blocked.blocked_by
WHERE blocked.role_id = 6
  AND blocked.blocked_at IS NOT NULL
ORDER BY blocked.blocked_at DESC;
"""

df = pd.read_sql(query, engine, parse_dates=["blocked_at"])
print(df)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

if df.empty:
    print("No blocked clients in this dataset.")
else:
    reason_counts = df["blocked_reason"].fillna("(no reason given)").value_counts()
    reason_counts.plot(kind="barh", color="#E45756")
    plt.gca().invert_yaxis()
    plt.title("Blocked clients by reason")
    plt.xlabel("Blocked clients")
    plt.ylabel("")
    plt.tight_layout()
    plt.show()
```

```python
if df.empty:
    print("No blocked clients in this dataset.")
else:
    admin_counts = df["blocked_by_admin"].fillna("(unknown admin)").value_counts()
    admin_counts.plot(kind="barh", color="#72B7B2")
    plt.gca().invert_yaxis()
    plt.title("Blocks issued per admin")
    plt.xlabel("Clients blocked")
    plt.ylabel("")
    plt.tight_layout()
    plt.show()
```
