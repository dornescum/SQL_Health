Select all users

```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
import pandas as pd

load_dotenv()

engine = create_engine(
    f"mysql+pymysql://{os.getenv('DB_USER')}:{os.getenv('DB_PASS')}"
    f"@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)
```

```python
df = pd.read_sql("SELECT id, name, surname, sex, country, town, role_id, is_active FROM users", engine)
df.head()
print(df)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

sex_counts = df["sex"].fillna("Unknown").value_counts()
sex_counts.plot(kind="bar", color=["#4C78A8", "#F58518", "#54A24B"])
plt.title("Users by sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
