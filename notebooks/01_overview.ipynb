# 01 — Overview

Connects to the `oloproteic` MySQL clone and pulls the patient/cardio master views.

```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
import pandas as pd

load_dotenv()

engine = create_engine(
    f"mysql+pymysql://{os.getenv('DB_USER')}:{os.getenv('DB_PASS')}"
    f"@{os.getenv('DB_HOST')}:{os.getenv('DB_PORT')}/{os.getenv('DB_NAME')}"
)
```

```python
df = pd.read_sql("SELECT * FROM view_ai_patient_master", engine)
df.head()
```
