# 01 — Overview

Connects to the `oloproteic` MySQL clone via the read-only `analyst` account and pulls the patient/cardio master views.

```python
import os
from dotenv import load_dotenv
from sqlalchemy import create_engine
import pandas as pd

load_dotenv()

engine = create_engine(
    f"mysql+pymysql://{os.getenv('DB_ANALYST_USER')}:{os.getenv('DB_ANALYST_PASSWORD')}"
    f"@{os.getenv('DB_ANALYST_HOST')}:{os.getenv('DB_ANALYST_PORT')}/{os.getenv('DB_ANALYST_NAME')}"
)
```

```python
df = pd.read_sql("SELECT * FROM view_ai_patient_master", engine)
df.head()
```
